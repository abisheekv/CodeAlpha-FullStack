```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import files

np.random.seed(0)

dates = pd.date_range(start='2022-01-01', end='2022-12-31', freq='D')
products = ['Product A', 'Product B', 'Product C']

sales_data = pd.DataFrame({
    'Date': np.random.choice(dates, 100),
    'Product': np.random.choice(products, 100),
    'Sales': np.random.randint(100, 1000, 100)
})

sales_data.to_csv('sales_data.csv', index=False)
files.download('sales_data.csv')
```

```python
DF1 = pd.read_csv('sales_data.csv')

#DF2= pd.read_csv("/content/train.csv")
```

| **Function**     | **Syntax**                                   | **Explanation**                                                    | **Example Code**                                               |
|------------------|----------------------------------------------|--------------------------------------------------------------------|---------------------------------------------------------------|
| **IMPORT**       | `import pandas as pd`                       | Imports the Pandas library using the standard alias `pd`.          | `import pandas as pd`                                          |
| **READ_CSV**     | `pd.read_csv(filename)`                      | Reads data from a CSV file.                                        | `df = pd.read_csv('data.csv')`                                 |
| **READ_TABLE**   | `pd.read_table(filename)`                    | Reads data from a delimited text file (like TSV).                  | `df = pd.read_table('data.tsv')`                               |
| **READ_EXCEL**   | `pd.read_excel(filename)`                    | Reads data from an Excel file.                                      | `df = pd.read_excel('data.xlsx')`                              |
| **READ_SQL**     | `pd.read_sql(query, connection_object)`      | Reads data from a SQL table/database using a query.                | `df = pd.read_sql('SELECT * FROM table_name', conn)`           |
| **READ_JSON**    | `pd.read_json(json_string)`                  | Reads from a JSON formatted string, URL, or file.                  | `df = pd.read_json('data.json')`                               |
| **READ_HTML**    | `pd.read_html(url)`                          | Parses an HTML URL, string, or file and extracts tables as DataFrames. | `dfs = pd.read_html('http://example.com/data.html')`          |
| **CLIPBOARD**    | `pd.read_clipboard()`                        | Reads the contents of your clipboard.                              | `df = pd.read_clipboard()`                                     |
| **DATAFRAME**    | `pd.DataFrame(dict)`                         | Creates a DataFrame from a dictionary; keys are column names, values are data as lists. | `df = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})`                |

```python
sns.countplot(x='Product', data=DF1)
plt.title('Product Count')
plt.show()
```

```python
DF1['Product'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(6,6)
)
plt.ylabel('')
plt.title('Product Sales Distribution')
plt.show()
```

```python
plt.hist(DF1['Sales'], bins=10)
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.title('Sales Distribution')
plt.show()
```

```python
sns.histplot(DF1['Sales'], kde=True)
plt.title('Sales Density Plot')
plt.show()
```

```python
plt.boxplot(DF1['Sales'])
plt.title('Sales Box Plot')
plt.show()

#kind can be 'line', 'bar', 'barh', 'hist', 'box', 'kde', 'density', 'area', 'scatter', 'pie', 'hexbin
```

```python
sns.scatterplot(x='Date', y='Sales', data=DF1)
plt.title('Sales over Time')
plt.show()
```

```python
sns.scatterplot(x='Date', y='Sales', hue='Product', data=DF1)
plt.title('Sales by Product Over Time')
plt.show()
```

```python
plt.boxplot(DF1['Sales'])
plt.title('Box Plot of Sales')
plt.ylabel('Sales')
plt.show()
```

```python
sns.scatterplot(x="Date", y="Sales", hue="Product", data=DF1)
plt.title("Sales over Time by Product")
plt.show()
```

```python
pd.crosstab(DF1['Product'], DF1['Sales'])
```

```python
DF1['Sales_Level'] = pd.cut(
    DF1['Sales'],
    bins=[0, 300, 600, 1000],
    labels=['Low', 'Medium', 'High']
)
```

```python
sns.heatmap(
    pd.crosstab(DF1['Product'], DF1['Sales_Level']),
    annot=True,
    cmap='Blues'
)
plt.title('Product vs Sales Level Heatmap')
plt.show()
```

| Plot Type   | Scenario                                                | Number of Variables | Suitable Variable Types           |
|-------------|---------------------------------------------------------|---------------------|------------------------------------|
| line        | Visualizing trends or changes over time (e.g., stock prices, sales) | 1 (for simple) or 2 (for comparing)  | Continuous (numeric, time series) |
| bar         | Comparing categories or discrete groups (e.g., sales by region) | 1 (for simple) or 2 (for grouped comparison) | Categorical (x) and continuous (y) |
| barh        | Similar to bar plot, but with horizontal bars for better readability | 1 (for simple) or 2 (for grouped comparison) | Categorical (x) and continuous (y) |
| hist        | Visualizing the distribution of a single continuous variable | 1                   | Continuous (numeric)              |
| box         | Showing the spread and outliers in data (e.g., income distribution) | 1                   | Continuous (numeric)              |
| kde         | Estimating the probability density function of a continuous variable | 1                   | Continuous (numeric)              |
| area        | Displaying cumulative data (stacked values over time) | 1 (for simple) or 2+ (for multiple series) | Continuous (numeric, time series) |
| scatter     | Showing relationships between two continuous variables (e.g., height vs. weight) | 2                   | Continuous (numeric)              |
| pie         | Showing proportions of a whole (e.g., market share)    | 1                   | Categorical (for labels) and continuous (for values) |
| hexbin      | Showing the relationship between two continuous variables in binned form (for large datasets) | 2                   | Continuous (numeric)              |
